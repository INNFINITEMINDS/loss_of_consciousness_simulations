```python
%reload_ext autoreload
%autoreload 2

import sys
for p in ['../simulations/', '../utils/']:
    if p not in sys.path:
        sys.path.append(p)

from brian2 import *
from plotlib import *
from tqdm import tqdm

import experiment_connectome_ping as ex10
import pickle
import scipy.stats as stats
```

```python
n_mod = 998
#n_mod = 100
start, end = 1000, 5000
dt, shift = 5, 5
```

```python

def mapper(modulation):
    return ex10.run_experiment(
        n_mod=n_mod,
        with_thalamus=True,
        thalamus_modulation=modulation,
        inter_scaling=50,
    )

p = Pool(4)
modulation = np.linspace(0, 1, 4)
data_array = p.map(mapper, modulation)
p.close()
```

```python
len(data_array)
```

```python
lzs = []
lzs = np.zeros((len(data_array), n_mod))
n_steps = float(end - start) / shift

for idx, d in enumerate(data_array):
    sys.stdout.write("{}... ".format(idx))
    sys.stdout.flush()
    lzs[idx,:] = get_lz_comp(d, start, end, dt, shift)
lzs *= np.log(n_steps) / n_steps
```

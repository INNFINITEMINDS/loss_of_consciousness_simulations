```python
#%matplotlib notebook
import scipy.io as spio
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from operator import itemgetter
from scipy.spatial import ConvexHull
import pandas as pd


plt.rc('axes', titlesize=25)
```

```python
CIJ = spio.loadmat('data/Conectoma.mat')['CIJ_fbden_average']
XYZ = spio.loadmat('data/coords_sporns_2mm.mat')['coords_new']
N = len(XYZ)
```

```python
xs, ys, zs = zip(*XYZ)
print("Length: {}".format(max(ys) - min(ys)))
print("Width: {}".format(max(xs) - min(xs)))
print("Height: {}".format(max(zs) - min(zs)))
min_x, max_x = min(xs), max(xs)
min_y, max_y = min(ys), max(ys)
min_z, max_z = min(zs), max(zs)
diam_y = max_y - min_y
diam_x = max_x - min_x
diam_z = max_z - min_z
diameter = max(
    diam_y, diam_x, diam_z
)

y0 = int(min_y - (diameter - diam_y) / 2.0)
y1 = int(min_y + diameter)
x0 = int(min_x - (diameter - diam_x) / 2.0)
x1 = int(x0 + diameter)
z0 = int(min_z - (diameter - diam_z) / 2.0)
z1 = int(z0 + diameter)
print(x0, x1, y0, y1, z0, z1)

#x0, x1 = 0, 90
#y0, y1 = 0, 90
#z0, z1 = 0, 90
    
```

```python
#point = 329
point = 278
point = 90
cmap = 'winter'
x0, y0, z0 = XYZ[point]
color_grad = np.zeros(N)
for i, xyz in enumerate(XYZ):
    x, y, z = xyz
    color_grad[i] = np.sqrt((x - x0) ** 2 + (y - y0) ** 2 + (z - z0) ** 2)
    
# Convex Hull
hull = ConvexHull(XYZ)

xs, ys, zs = map(np.array, zip(*XYZ))

#fig = plt.figure(figsize=plt.figaspect(1)*4) #Adjusts the aspect ratio and enlarges the figure (text does not enlarge)
#ax = fig.gca(projection='3d')
#fig = plt.figure()
fig = plt.figure(figsize=(15, 15))
fig.subplots_adjust(hspace=0, wspace=0)
gs1 = gridspec.GridSpec(2, 2)
gs1.update(wspace=0, hspace=0)#0.025, hspace=0.05)

ax = plt.subplot(gs1[0], projection='3d')
#ax = fig.add_subplot(221, projection='3d')
point_size = 5

mask = np.arange(N) != point
ax.scatter(xs, ys, zs, s=point_size, cmap=cmap, c=xs)
#ax.scatter(xs[mask], ys[mask], zs[mask], c=color_grad[mask], s=5, cmap='viridis')
#ax.scatter(xs[point], ys[point], zs[point], c='red')
#for simplex in hull.simplices:
#    ax.plot(XYZ[simplex, 0], XYZ[simplex, 1], XYZ[simplex, 2], 'k-', lw=0.1)
ax.set_title('Left view')
ax.view_init(10, 0)

ax2 = plt.subplot(gs1[1], projection='3d')
#ax2 = fig.add_subplot(222, projection='3d')
ax2.scatter(xs, ys, zs, s=point_size, cmap=cmap, c=xs)
ax2.set_title('Front view')
ax2.view_init(10, -90)

ax3 = plt.subplot(gs1[2], projection='3d')
#ax3 = fig.add_subplot(223, projection='3d')
ax3.scatter(xs, ys, zs, s=point_size, cmap=cmap, c=xs)
ax3.set_title('Right view')
ax3.view_init(10, 180)

ax4 = plt.subplot(gs1[3], projection='3d')
#ax4 = fig.add_subplot(224, projection='3d')
ax4.scatter(xs, ys, zs, s=point_size, cmap=cmap, c=xs)
ax4.set_title('Top view')
ax4.view_init(85, -90)

ax.set_xticklabels([])
ax2.set_yticklabels([])
ax3.set_xticklabels([])
ax4.set_zticklabels([])

#plt.gca().set_aspect('equal', adjustable='box')
for axs in [ax, ax2, ax3, ax4]:
    axs.axis('equal')
#     axs.auto_scale_xyz( #         [x0, x1], [y0, y1], [z0, z1] # #        [z0, z1], [z0, z1], [z0, z1] # #        [0, 90], [0, 90], [0, 90] #     )
    
#      axs.auto_scale_xyz(
#          [min(ys), max(ys)], [min(ys), max(ys)], [min(ys), max(ys)])
    axs.set_xticklabels([])
    axs.set_yticklabels([])
    axs.set_zticklabels([])
    axs.set_axis_off()
plt.tight_layout()
#plt.autoscale(False)
plt.savefig('figures/connectome_3D.png')
plt.show()
```

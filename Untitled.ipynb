# Modular connectome - Scaling 

```python
import experiment_10_ping_modular_connectome as ex10
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from brian2 import *
import power_spectral_density as psd
from IPython.display import Image
```

### Running Experiment with scaling = 100

```python
N = 1000
duration = 5000
conn = 0.1
scaling = 100
data1 = ex10.run_experiment(N, duration, conn, scaling, verbose=True)
```

### Running Experiment with scaling = 50

```python
N = 1000
duration = 5000
conn = 0.1
scaling = 50
data2 = ex10.run_experiment(N, duration, conn, scaling, verbose=True)
```

### Running Experiment with scaling = 10

```python
N = 1000
duration = 5000
conn = 0.1
scaling = 10
data3 = ex10.run_experiment(N, duration, conn, scaling, verbose=True)
```

```python

N = 1000
duration = 5000
conn = 0.1
scaling = 1
data4 = ex10.run_experiment(N, duration, conn, scaling, verbose=True)
```

## LZ Complexity Distributions

```python

nbins = 100
datasets = reversed([data1, data2, data3, data4])
for d, scale in zip(datasets, reversed([100, 50, 10, 1])):
    lz = d['lz_comp']
    n_steps = d['n_steps']
    norm = np.log(n_steps) / n_steps
    #y, bin_edges = np.histogram(lz * norm, bins=nbins)
    #bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    #plt.plot(bin_centers, y, label='Scaling {}'.format(scale))
    plt.hist(lz * norm, label='Scaling {}'.format(scale))
    
plt.xlim([0, 1])
plt.title('LZ Complexity Distribution')
plt.legend()
plt.show()
```

```python
x1, y1, n_mod1, m_v1 = [data1[k] for k in ['X', 'Y', 'N_MOD', 'M_V']]
x2, y2, n_mod2, m_v2 = [data2[k] for k in ['X', 'Y', 'N_MOD', 'M_V']]
x3, y3, n_mod3, m_v3 = [data3[k] for k in ['X', 'Y', 'N_MOD', 'M_V']]
x4, y4, n_mod4, m_v4 = [data4[k] for k in ['X', 'Y', 'N_MOD', 'M_V']]
```

## Raster Plots

```python
mask = np.logical_and.reduce((x1 >= 1000, x1 < 1500, y1 < 1000))
plt.plot(x1[mask], y1[mask], '.')
plt.title('Weight scaling: 100')
plt.xlabel('Simulation Time')
plt.ylabel('Neuron Index')
plt.show()
```
